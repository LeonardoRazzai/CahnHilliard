```python
from CahnHilliard import *
```

**Setup the spatial grid**

```python
L = 40
N = 400
dx = L / N
dy = L / N
```

**Define the initial concentration field $c_0$**

```python
mean = 0.0 # must be less than 0.69
c0 = np.zeros((N, N)) + mean

amp = 0.01
c0 = c0 + amp * np.random.randn(N, N)
```

**Define diffusivity and interfacial parameter**

```python
D = dx**2 * 3 # diffusivity cm**2 / s

gamma = 500
beta = N / dx / gamma # in this way fastest growing wavelength is 2*pi * L/gamma
a = dx**4 * beta**2 # fastest growing componet is 1/sqrt(a)
```

**Compute the solution up to tmax = 5**

```python
dt = 0.01
tmax = 5
time = np.arange(0, tmax, dt)

CH_system = Sol_CahnHilliard(L, N, D, a)
CH_system.compute_sol(c0, time)
```

```python
CH_system.set_step(10)
```

```python
CH_system.ComputeFT()
```
